```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

# Gift Wrapping 2D 

Esse primeiro algoritmo é printando cada passo, serviu de auxilio para fazer, se quiser o algoritmo que te da o resultado pronto so pular para o proximo

##### 1º

```python
Pontos = int(input("Qual é o número de pontos? "))
P = [(np.random.randint(0,200),np.random.randint(0,200)) for i in range(Pontos)]
t0 = time.time()

def VirarDireita(cord1,cord2,cord3):
    if (cord3[1]-cord1[1])*(cord2[0]-cord1[0]) >= (cord2[1]-cord1[1])*(cord3[0]-cord1[0]):
        return True
    return False

def GiftWrapping(P):
    P.sort()
    #print(P,"\n Coordenadas na ordem")
    P = np.array(P) 
    Top = [P[0],P[1]]
    for i in range(2,len(P)):
        Top.append(P[i])
        while len(Top) > 2 and VirarDireita(Top[-1],Top[-2],Top[-3]):
            del Top[-2]
        F = np.array(Top)
        plt.plot(F[:,0],F[:,1])
        plt.plot(P[:,0],P[:,1],".r")
        plt.show()
    Baixo = [P[-1],P[-2]]
    for i in range(len(P)-3,-1,-1):
        Baixo.append(P[i])
        while len(Baixo) > 2 and VirarDireita(Baixo[-1],Baixo[-2],Baixo[-3]):
            del Baixo[-2]
        L = np.array(Top + Baixo)
        #plt.clf()
        plt.plot(P[:,0],P[:,1],".r")
        plt.plot(L[:,0],L[:,1])
        plt.show()


GiftWrapping(P)

t1 = time.time()
total = t1-t0
print(total)
```
